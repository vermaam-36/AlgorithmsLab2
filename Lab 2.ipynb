## 1. Decision Rule


Given a 100x100 2D plane. For any polygon with points {p1, p2...pn} n >= 3 and point m {x,y} 0 <= x <= 100, 0 <= y <= 100.  

Let 4 lines be drawn outward from point m, one horizontal vector represented as a line.  

If the vector touchs lines created by the polygon points {p1, p2...pn} and odd amount of times then the point m lies within the polygon else point m lies outside the polygon  
  
Run Time Analysis:  
The algorithm runs in O(n) time as a worst case dependent on the number of points the polygon has if there is no polygon then we see a best case complexity of O(1) as the for loop will never be tripped. The line intersection method called inside the algorithm has a complexity of O(1) best and worst case.

### Find max and min, check if intersectio npoint is within max and min

```python
def check_inside(polygon, st_pt):
    x_max = max(polygon)[0] # + 1
    horizontal_line= [0, 1, st_pt[1]] # + 1
    intersections= 0 # + 1

    for i in range(0, len(polygon)): #1 + n(1 
        point1=polygon[i] # + 1
        if(i != len(polygon) - 1): # + 1 + b(1)
            point2=polygon[i+1]
        else: # + 1 - b(1)
            point2=polygon[0]
        
        polygon_line= create_line(point1, point2) # + 1 (create line runs O(1) time)
        intersection_point=findIntersection(polygon_line, horizontal_line, st_pt) # + 1 + O(1) <- from findIntersection()

       
        if(intersection_point is not None): # + 1 + b(
             
            if(intersection_point[0]==point2[0] and intersection_point[1]==point2[1]): # 1 + b(3)
                intersections+=0
                #This is the double count case
            elif(intersection_point[0]>st_pt[0]): # + 1 + b(3
                if(x_max >= intersection_point[0]): # + 1 + b(2)
                    intersections+=1

    if(intersections%2 == 1): # + 1 + b(1)
        return True
    

    return False # + 1
#T(n) = 1 + 1 + 1 + 1 + 1 + n(1 + 1 + b(1) + 1 - b(1) + 1 + 1 + O(1) + 1 + 1 + 1 + 1 + b(1 + b(3) + 1 + b(1)) + 1 + 1) + 1 + 1 + b(2) + 1
#T(n) = 10 + 19n in O(n) worst and O(1) if there is no polygon to check
```

```python
def create_line(pt1, pt2):

    x1 = pt1[0]
    y1 = pt1[1]

    x2 = pt2[0]
    y2 = pt2[1]

    a = y2 - y1
    b = x1 - x2

    c = (a * x1) + (b * y1)
    return np.array([a, b, c])
```

```python
def findIntersection(poly, checker, st_pt):
    try: #1 + a(
        if(poly[0] == 0 and poly[1] % checker [1] == 0 and poly[2] % checker[2] == 0): # + 1 + b(1)
            return None
        
        m_list = [[poly[0], poly[1]], [checker[0], checker[1]]] # + 1
        A = np.array(m_list) # + 1
        inv_A = np.linalg.inv(A) # + 1
#         print(inv_A)
        B = np.array([poly[2], checker[2]]) # + 1
        X = np.linalg.inv(A).dot(B) # + 1
#         print(X)
        return X # + 1
    except: # + 1)
        return # + 1       T(n) = 1 + a(1 + b(1) + 1 + 1 + 1 + 1 + 1 + 1 + 1) + 1 = 8 = O(1) worst and best case
```

## Complex Polygon Test

```python
def draw_polygon(n):
    # https://stackoverflow.com/questions/45831084/creating-a-polygon-in-python
    sort_tups = []
    x = np.random.randint(0,50,n)
    y = np.random.randint(0,50,n)

    ##computing the (or a) 'center point' of the polygon
    center_point = [np.sum(x)/n, np.sum(y)/n]

    angles = np.arctan2(x-center_point[0],y-center_point[1])

    ##sorting the points:
    sort_tups = sorted([(i,j,k) for i,j,k in zip(x,y,angles)], key = lambda t: t[2])

    ##making sure that there are no duplicates:
    if len(sort_tups) != len(set(sort_tups)):
        raise Exception('two equal coordinates -- exiting')

    x,y,angles = zip(*sort_tups)
    x = list(x)
    y = list(y)
    
    return [(i, j) for i, j in zip(x, y)]
```

```python
coord = draw_polygon(15)
st_pt = np.rint(np.mean(coord, axis=0))
coord.append(coord[0]) #repeat the first point to create a 'closed loop'

xs, ys = zip(*coord) #create lists of x and y values

plt.figure()
plt.plot(xs,ys)
plt.plot(st_pt[0], st_pt[1], 'ro')
plt.show()

check_inside(coord[:-1], st_pt)
```

## Benchmarking

```python
def benchmark(algorithm, polygon, st_pt):
    # Setup
    start_time = time.perf_counter()
    
    # Do benchmarking
    algorithm(polygon, st_pt)
    
    end_time = time.perf_counter()
    elapsed = end_time - start_time
    return elapsed
```

```python
benchmark(check_inside, coord, st_pt)
```
